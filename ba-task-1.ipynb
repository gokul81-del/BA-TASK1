```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "Price": [20, 25, 30, 35, 40],
    "Bottles Sold": [120, 110, 90, 80, 70]
}
```

```python
df = pd.DataFrame(data)
```

```python
print("Original Dataset:")
print(df)
```

```python
df["Revenue"] = df["Price"] * df["Bottles Sold"]
```

```python
print("\nUpdated Dataset:")
print(df)
```

```python
highest_revenue = df["Revenue"].max()

best_prices = df[df["Revenue"] == highest_revenue]["Price"].tolist()

print("\nHighest Revenue: ₹", highest_revenue)
print("Price(s) with Highest Revenue: ₹", best_prices)
```

```python
plt.figure(figsize=(6,4))
plt.bar(df["Price"].astype(str), df["Revenue"])
plt.title("Price vs Revenue")
plt.xlabel("Price (₹)")
plt.ylabel("Revenue (₹)")
plt.show()
```

```python
print("\nBusiness Conclusion:")
print("The highest revenue is ₹2800.")
print("Both ₹35 and ₹40 generate the highest revenue.")
print("The shop owner can choose ₹35 because it earns the maximum revenue while keeping the selling price lower than ₹40.")
```
